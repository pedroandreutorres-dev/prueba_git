## Part 1: Initial Setup

1. Create a new folder on your Desktop named `learning-git`.
2. Open that folder in VS Code.
3. Open the terminal (`Ctrl + ` ` ` or `Ctrl + √±`).
4. Initialize the repository:
   ```bash
   git init
   ```
5. Check that a hidden folder `.git` has been created (use `ls -a`).

## Part 2: The Commit Cycle

1. Create a file named `main.py` and write `print("Hello Git")` inside.
2. Check the status:
   ```bash
   git status
   ```
   *(You should see main.py in red - Untracked)*

3. Add the file to the Staging Area:
   ```bash
   git add main.py
   ```

4. Check status again (`git status`). It should be green.

5. Create the commit:
   ```bash
   git commit -m "Initial commit: created main script"
   ```

6. Check the history:
   ```bash
   git log
   ```

## Part 3: Making Changes

1. Modify `main.py`. Add a second line: `print("This is a change")`.
2. Create a new file `utils.py`.
3. Run `git status`. You will see one "modified" file and one "untracked" file.
4. Add everything at once:
   ```bash
   git add .
   ```
5. Commit:
   ```bash
   git commit -m "Added utils and updated main"
   ```

## Part 4: The Ignore File

1. Create a file named `secret_passwords.txt`.
2. We don't want this in Git. Create a file named `.gitignore`.
3. Inside `.gitignore`, write:
   ```text
   *.txt
   ```
4. Run `git status`. `secret_passwords.txt` should **NOT** appear. Only `.gitignore` should appear.
5. Commit the `.gitignore` file.

## Part 5: GitHub (The Cloud)

1. Go to GitHub.com and create a **New Repository** (empty, no README).
2. Copy the URL (HTTPS).
3. Link your local repo to GitHub:
   ```bash
   git remote add origin <PASTE_YOUR_URL_HERE>
   ```
4. Rename your branch to main (if it's not already):
   ```bash
   git branch -M main
   ```
5. Push your code:
   ```bash
   git push -u origin main
   ```
6. Refresh GitHub page. Your code should be there!
