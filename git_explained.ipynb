## 1. What is Git?
Git is a **version control system**. It allows you to track changes in your files, travel back in time to previous versions, and work with other people without overwriting each other's code.

Think of Git as a **Time Machine** for your project.

---

## 2. First Setup (One time only)
Before using Git, you must introduce yourself. Git prints this info in every "save" (commit) you make.

Open your terminal (Git Bash in Windows) and run:

```bash
git config --global user.name "Your Name"
git config --global user.email "your.email@example.com"
```

Check your config:
```bash
git config --list
```

---

## 3. The Local Workflow (The 3 Stages)

Git has three main areas. Understanding this is the key to mastering Git.

1.  **Working Directory:** Where you are currently writing code (your files).
2.  **Staging Area:** A waiting room. You choose which files you want to save.
3.  **Repository:** The permanent database of versions (commits).

### Diagram:
```text
Working Dir  --[git add]-->  Staging Area  --[git commit]-->  Repository
```

### Key Commands

| Command | Description |
| :--- | :--- |
| `git init` | Turns the current folder into a Git repository. |
| `git status` | **Crucial.** Tells you what is going on (modified files, staged files). |
| `git add filename` | Moves a file to the Staging Area. |
| `git add .` | Moves **ALL** changed files to the Staging Area. |
| `git commit -m "msg"` | Saves a snapshot of the Staging Area permanently. |
| `git log` | Shows the history of commits (the timeline). |

---

## 4. Branches (Parallel Universes)

Branches allow you to work on new features without breaking the main code. Ideally, you never work directly on `main`.



| Command | Description |
| :--- | :--- |
| `git branch` | Lists all local branches. |
| `git branch new-feature` | Creates a new branch named 'new-feature'. |
| `git switch new-feature` | Switches to that branch (older Git uses `checkout`). |
| `git merge new-feature` | Merges 'new-feature' into the current branch. |

---

## 5. Remote Repositories (GitHub/GitLab)

To share your code or have a backup in the cloud.

| Command | Description |
| :--- | :--- |
| `git clone <url>` | Downloads a repository from the internet to your PC. |
| `git remote add origin <url>` | Links your local folder to a GitHub repo. |
| `git push origin main` | Uploads your commits to the cloud. |
| `git pull origin main` | Downloads new commits from the cloud. |

---

## 6. The `.gitignore` file

There are files we **never** want to upload to Git:
* Passwords / API Keys.
* Temporary files (`__pycache__`, `.DS_Store`).
* Virtual environments (`.venv`, `env`).

To ignore them, create a file named `.gitignore` and write the names of the files/folders inside.
